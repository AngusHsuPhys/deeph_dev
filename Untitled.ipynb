```python
# Initialize an empty dictionary to store line positions and contents
overlap_matrix_lines = {}

# Open the file and search for lines containing the specific phrase
with open('../example2/work_dir/dataset/raw/0/HS.out', 'r') as file:
    lines = file.readlines()
    for index, line in enumerate(lines):
        if line.startswith("Overlap matrix with position operator"):
            # Store the line and the next five lines in the dictionary
            overlap_matrix_lines[index + 1] = lines[index:index + 6]

# Print the dictionary containing the line positions and contents
for position, content in overlap_matrix_lines.items():
    print(f"Starting Line {position}:")
    for line in content:
        print(len(line.split()))
        print(line.split())
```
