```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import cumtrapz

# Read the data from the file
filename = './deeph_dev/openmx.DOS.Gaussian'  # Replace with your filename
data = np.loadtxt(filename)

# Extract the first and second columns
energy = data[:, 0]
dos = data[:, 1]

# Specify the Fermi level (in eV)
fermi_level = -3.670901831905665# Replace with your actual Fermi level

# Calculate the cumulative integral of the DOS (occupation number)
occupation_number = cumtrapz(dos, energy, initial=0)

# Find the occupation number at the Fermi level
index_fermi = np.searchsorted(energy, fermi_level)
occupation_at_fermi = occupation_number[index_fermi]

# Plot the DOS
plt.plot(energy, dos, color='black')
plt.xlabel('Energy (eV)')
plt.ylabel('Density of States (DOS)')
plt.title('Density of States (DOS) Plot')
plt.grid(True)
plt.axvline(x=fermi_level, color='red', linestyle='--', label='Fermi Level')
plt.legend()

# Plot the occupation number
plt.figure()
plt.plot(energy, occupation_number, color='blue')
plt.xlabel('Energy (eV)')
plt.ylabel('Occupation Number')
plt.title('Occupation Number Plot')
plt.grid(True)
plt.axvline(x=fermi_level, color='red', linestyle='--', label=f'Fermi Level: {occupation_at_fermi:.2f}')
plt.legend()

plt.show()

# Print the occupation number at the Fermi level
print(f"Occupation number at the Fermi level ({fermi_level} eV): {occupation_at_fermi}")
```
