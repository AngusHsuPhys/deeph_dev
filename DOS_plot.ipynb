```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import cumtrapz

# Read the data from the file
filename = '/home/t.hsu/example2/work_dir/inference/5_4/dos.dat'  # Replace with your filename
data = np.loadtxt(filename)

# Extract the first and second columns
energy = data[:, 0]
dos = data[:, 1]

# Specify the Fermi level (in eV)
fermi_level = -3.670901831905665# Replace with your actual Fermi level

# Calculate the cumulative integral of the DOS (occupation number)
occupation_number = cumtrapz(dos, energy, initial=0)

# Find the occupation number at the Fermi level
index_fermi = np.searchsorted(energy, fermi_level)
occupation_at_fermi = occupation_number[index_fermi]

# Plot the DOS
plt.plot(energy, dos, color='black')
plt.xlabel('Energy (eV)')
plt.ylabel('Density of States (DOS)')
plt.title('Density of States (DOS) Plot')
plt.grid(True)
plt.axvline(x=fermi_level, color='red', linestyle='--', label='Fermi Level')
plt.legend()

# Plot the occupation number
plt.figure()
plt.plot(energy, occupation_number, color='blue')
plt.xlabel('Energy (eV)')
plt.ylabel('Occupation Number')
plt.title('Occupation Number Plot')
plt.grid(True)
plt.axvline(x=fermi_level, color='red', linestyle='--', label=f'Fermi Level: {occupation_at_fermi:.2f}')
plt.legend()

plt.show()

# Print the occupation number at the Fermi level
print(f"Occupation number at the Fermi level ({fermi_level} eV): {occupation_at_fermi}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import cumtrapz

# Define the range of file indices
file_indices = range(0, 81)

# Initialize lists to hold data for all files
all_energy = []
all_dos = []
all_occupation_numbers = []
fermi_levels = []
occupation_at_fermis = []

# Specify the Fermi level (in eV)
Chem_potential = -0.13490316537390
Hart2eV = 27.2114
fermi_level = Chem_potential * Hart2eV

for i in file_indices:
    filename = f'/home/t.hsu/DeepRFnet/DeepDOS/work_dir/dataset/raw/{i}/openmx.DOS.Gaussian'
    data = np.loadtxt(filename)

    # Extract the first and second columns
    energy = data[:, 0]
    dos = data[:, 1]

    # Calculate the cumulative integral of the DOS (occupation number)
    occupation_number = cumtrapz(dos, energy, initial=0)

    # Find the occupation number at the Fermi level
    index_fermi = np.searchsorted(energy, fermi_level)
    occupation_at_fermi = occupation_number[index_fermi]

    # Store data
    all_energy.append(energy)
    all_dos.append(dos)
    all_occupation_numbers.append(occupation_number)
    fermi_levels.append(fermi_level)
    occupation_at_fermis.append(occupation_at_fermi)

# Create subplots
fig, (ax1, ax2) = plt.subplots(2, 1, sharex=True, figsize=(10, 8))

# Plot all DOS
for energy, dos in zip(all_energy, all_dos):
    ax1.plot(energy, dos, alpha=0.5)
ax1.set_ylabel('Density of States (DOS)')
ax1.set_title('Density of States (DOS) and Occupation Number Plots')
ax1.grid(True)
# ax1.axvline(x=fermi_level, color='red', linestyle='--', label='Fermi Level')
ax1.legend()

# Plot all occupation numbers
for energy, occupation_number in zip(all_energy, all_occupation_numbers):
    ax2.plot(energy, occupation_number, color='blue', alpha=0.5)
ax2.set_xlabel('Energy (eV)')
ax2.set_ylabel('Occupation Number')
ax2.grid(True)
ax2.axvline(x=fermi_level, color='red', linestyle='--', label='Fermi Level')
ax2.legend()

# Show the plots
plt.tight_layout()
plt.show()

# Print the occupation number at the Fermi level for each file
for i, occupation_at_fermi in enumerate(occupation_at_fermis):
    print(f"Occupation number at the Fermi level for file {i} ({fermi_level} eV): {occupation_at_fermi}")
```
