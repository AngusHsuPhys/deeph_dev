```python
# Initialize an empty dictionary to store line positions and contents
overlap_matrix_lines = {}

# Open the file and search for lines containing the specific phrase
with open('../example2/work_dir/pos/5_4/HS3.out', 'r') as file:
    index_list = []
    lines = file.readlines()
    for index, line in enumerate(lines):
    
        if line.startswith("Overlap matrix with position operator x"):
            index_list.append(index)
        elif line.startswith("Overlap matrix with position operator y"):
            index_list.append(index)
```

```python
non_zero_lines = {}

# Define the start and end indices for the search
start_index = index_list[0]
end_index = index_list[1]

# Open the file and search for lines with non-zero elements within the specified index range
with open('../example2/work_dir/pos/5_4/HS3.out', 'r') as file:
    # Iterate through the file line by line
    for current_index, line in enumerate(file):
        # Only process lines within the specified range
        if start_index <= current_index <= end_index:
            if line.startswith('global index='):
                continue
            # Check if the line contains any non-zero elements
            if any(num not in {'0.0000000', '-0.0000000'} for num in line.split()):
                non_zero_lines[current_index] = line.strip()  # Store the line content without leading/trailing whitespace
```
